```python
import numpy as np
import pandas as pd
import os
```

## 1. Sneak peak of an image.

```python
# load and display an image with Matplotlib
from matplotlib import image
from matplotlib import pyplot
# load image as pixel array
data = image.imread('mnistTask3/mnistTask/0/image_000007.jpg')
# summarize shape of the pixel array
print(data.dtype)
print(data.shape)
# display the array of pixels as an image
pyplot.imshow(data)
pyplot.show()
```

## 2. Store data in DataFrame.

```python
#Intialize empty dataframe.
width_req, height_req=[28,28]
dataframe={'label':[]}
for i in range(1,width_req*height_req+1):
    dataframe['pixel'+str(i)]=[]
df = pd.DataFrame(dataframe)
df
```

```python
for sample in os.listdir('mnistTask3/mnistTask/'): #For every folder
    for img in os.listdir('mnistTask3/mnistTask/'+sample): #For every image in the folder
        data_to_add = image.imread('mnistTask3/mnistTask/'+sample+'/'+img) #Read image
        data_to_add.resize(28*28) #Flatten into a row.
        data_to_add=np.insert(data_to_add,0,sample) #Add label to row
        df.loc[len(df.index)]=data_to_add #Add row to dataframe.
    print('Completed with '+sample+'/9')
```

```python
df.to_csv('preprocessed_dataset3.csv', index = False) #Store DataFrame as a csv file.
```
